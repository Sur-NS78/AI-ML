# Random Forest

Instead of using a single predictive model, we can combine the results of multiple models to achieve higher accuracy and better overall performance.

This is known as ensembling—the process of combining the predictions of multiple models on a dataset.

The base model trained in a Random Forest (RF) is a Decision Tree (Refer: [Decision-Tree.ipynb](Decision-Tree.ipynb)). Each decision tree is trained on a random sample of the dataset (a.k.a. bootstrap sampling), and each tree considers only a random subset of features during splitting.

This randomness in sampling and features ensures that each tree learns different patterns trees are less correlated with each other, reducing overfitting.

Each tree will grow using its sampled dataset and continue until it meets a stopping criterion (e.g., reaching a pure node, reaching the max_depth , or not having enough samples to split at the node).


## For Classification
Here, the RF conducts majority voting-based classification, meaning it outputs the class predicted by the majority of the trees.

## For Regression
The RF averages all the predictions from each tree to get the final overall prediction on the dataset.

$$ \hat y = \frac{1}{T} \sum_{t=1}^{T} h_t(x)$$

- $h_t(x)$ -> b-th tree prediction
- $T$  -> total no. of trees 

## OOB (Out of Bag) Error
Out-of-Bag error is an interbal validation method for the model for generalisation.
Because of we randomly sample the data from the dataset with replacement, some data points are left out of that sample— knnown as out-of-bag (OOB) samples for that tree.
For each data point, we use only the trees that did not see it during training to make predictions, and then calculate error of the prediction and by aggregating these errors across all data points we get the OOB score.



```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

data=load_breast_cancer()

X=data.data
y=data.target

X_tr,X_test,Y_tr,Y_test=train_test_split(X,y,test_size=0.25,random_state=42)
RF=RandomForestClassifier(n_estimators=100,random_state=42,oob_score=True)
RF.fit(X_tr,Y_tr)
pred=RF.predict(X_test)
print("Accuracy:", accuracy_score(Y_test, pred))    
print("Classification Report:\n", classification_report(Y_test, pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, pred))
```

```python

print("OOB Score:", RF.oob_score_)

```

```python

from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

california_housing=fetch_california_housing()
X=california_housing.data
y=california_housing.target
X_tr,X_test,Y_tr,Y_test=train_test_split(X,y,test_size=0.25,random_state=42)
RF=RandomForestRegressor(n_estimators=120,random_state=42,n_jobs=-1,oob_score=True)
RF.fit(X_tr,Y_tr)
pred=RF.predict(X_test)
print("Mean Squared Error:", mean_squared_error(Y_test, pred))
print("R2 Score:", r2_score(Y_test, pred))
```

```python

print("OOB R2 Score:", RF.oob_score_)
```
