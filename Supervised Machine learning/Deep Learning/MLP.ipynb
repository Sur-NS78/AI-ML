# Multilayer Perceptron (Neural Networks)
The most basic and powerful model architectures in domain of deep learning due to its flexibility and ability to learn complex non linear patterns. It serves as the core building block for advanced deep learning models like CNNs, LSTMs, RNNs and Transformers.

it is a model consisting of multiple layers of interconnected nodes to learn patterns from data. The layers consist of three types: 
The input layer which receives the data, The hidden layer(s) that transforms the data into a more useful representation and The output layer produces the final prediction.

Each layer is made up of units called neurons.


## Inside a neuron


Each neuron applies a linear transformation to its inputs and then passes the result through a non-linear activation function.

For an input vector $x$, the neuron computes:

$$
z = w^T x + b
$$

where:
- $w$ is the weight vector  
- $b$ is the bias  

This is followed by an activation function:

$$
a = \sigma(z)
$$
One commonly used activation function is the sigmoid function:
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

The activation function allows the network to model complex patterns.


## Forward propagation

An MLP processes data layer by layer. The output of one layer becomes the input to the next.

For a hidden layer $l$, the computation is:

$$
z^{(l)} = W^{(l)} a^{(l-1)} + b^{(l)}, \quad a^{(l)} = \sigma(z^{(l)})
$$

This process continues until the output layer produces the final prediction.


## Output layer and tasks

The choice of activation function in the output layer depends on the task our netwrok is goin to be uused for:

- regression → linear activation  
- binary classification → sigmoid  
- multiclass classification → softmax  


## Training the MLP

The model learns by minimizing a loss function using gradient-based optimization.

Loss functions include commonly used are mean squared error for regression cross-entropy loss for classification  

The training process involves:

_forward propagation to compute predictions  ->  loss computation -> backpropagation to compute gradients  -> updating weights using an optimizer_


## Backpropagation and optimisation


Each parameter in the network affects the final loss through multiple layers.  
Backpropagation computes these effects step by step using the chain rule, starting from the output layer and moving backward.

For the output layer, the error is computed as:

$$
\delta^{(L)} = \frac{\partial L}{\partial z^{(L)}}
$$

This term represents how much the output layer contributes to the loss.

For each hidden layer, the error is propagated backward:

$$
\delta^{(l)} = (W^{(l+1)})^T \delta^{(l+1)} \cdot \sigma'(z^{(l)})
$$

This means the model will take the error from the next layer, scale it using the weights adjust them using the derivative of the activation function 

Once we have $\delta^{(l)}$, we compute:

$$
\frac{\partial L}{\partial W^{(l)}} = \delta^{(l)} (a^{(l-1)})^T
$$

$$
\frac{\partial L}{\partial b^{(l)}} = \delta^{(l)}
$$

These gradients tell us how to update the parameters to reduce the loss.


### Parameter update

The parameters are updated using an optimizer such as SGD or Adam:

$$
W := W - \eta \frac{\partial L}{\partial W}, \quad
b := b - \eta \frac{\partial L}{\partial b}
$$

where $\eta$ is the learning rate.

This update gradually reduces the error over multiple iterations.



```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier,MLPRegressor
```

```python

from sklearn.datasets import load_digits


digits = load_digits()

X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=300, random_state=42)

mlp.fit(X_train, y_train)

accuracy = mlp.score(X_test, y_test)
print(f"Accuracy: {accuracy:.2f}")

y_pred = mlp.predict(X_test)
print(f"Predicted labels: {y_pred}")
print(f"True labels: {y_test}")
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler


housing = fetch_california_housing()


X_train, X_test, y_train, y_test = train_test_split(
    housing.data, housing.target, test_size=0.2, random_state=42
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#
mlp_regressor = MLPRegressor(hidden_layer_sizes=(64,32), max_iter=300, random_state=42)
mlp_regressor.fit(X_train_scaled, y_train)


r2_score = mlp_regressor.score(X_test_scaled, y_test)
print(f"R^2 Score: {r2_score:.2f}")
print(f"MSE: {mean_squared_error(y_test, mlp_regressor.predict(X_test_scaled)):.2f} ")
```
