#  The K-Nearest Neighbors algorithm
This is another simplistic model that classifies data points based on their distance from training data points.
It stores the entire training data, compares new data points with it and predicts the output based on the k nearest data points. this is why its known as a 'lazy' algorithm.

In terms of classification it predicts the class of the data sample based on the class of the k nearest points to the data sample.

For Regression, the prediction will be the mean of k nearest data point's values.

The k parameter will heavily influence the prediction of the model as it is the one deciding the amount of neighbors to consider for prediction decision

### The distance measures utilised

#### Euclidean
This is the straight line distance between the two data points in a Euclidean Space(2d), used majorly in knn for continuous variables.
$$\sqrt{ \sum_{i=1}^{n} (X^i_1-X^i_2)^2}$$
#### Hamming
We use this for categorical variables in the data

$$ \sum_{i=1}^{k} |X^i_1-X^i_2|$$

If both are same then distance is 0 , else it is 1

#### Manhattan
Manhattan distance is a way to measure how far apart two data points are by adding the absolute differences between their feature values.

It is similar to moving in a grid, where you can only move along horizontal and vertical paths.

$$
d(x_1, x_2) = \sum_{i=1}^{n} |X^i_1 - X^i_2|
$$

Here, $x_1$ and $x_2$ are the two data points, and $n$ is the number of features. For each feature, we take the difference between the two values, convert it to a positive value (absolute difference), and then add all these differences together.

Instead of measuring the shortest straight-line distance between two points, Manhattan distance works by adding the step-by-step differences across each feature. In simple terms, it calculates how far two points are by moving one feature at a time.

#### Minkowski

Instead of defining distance in just one fixed way, It uses a parameter that controls how the distance is calculated. By changing this parameter, we can obtain different types of distance measures.

The formula is:

$$
d(x_1, x_2) = \left( \sum_{i=1}^{n} |X^i_1 - X^i_2|^p \right)^{\frac{1}{p}}
$$

Here, $x_1$ and $x_2$ are the two data points, $n$ is the number of features, and $p$ is a parameter that determines the type of distance. For each feature, the difference is taken, raised to the power $p$, summed across all features, and then the final result is taken to the power of $\frac{1}{p}$.

The important idea is that different values of $p$ give different distance measures. 
##### Case examples:
 When $p = 1$, Minkowski distance becomes Manhattan distance.

When $p = 2$, it becomes Euclidean distance. As $p$ increases further,the distance becomes more sensitive to larger differences in individual features.



```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn .metrics import accuracy_score,classification_report,confusion_matrix

datast=load_wine()
X=datast.data
y=datast.target

Xtr,Xtest,ytr,ytest=train_test_split(X,y,test_size=0.3,random_state=42)

KnnClassifier=KNeighborsClassifier(n_neighbors=4)
KnnClassifier.fit(Xtr,ytr)  
pred=KnnClassifier.predict(Xtest)
print("Accuracy:", accuracy_score(ytest, pred))
print("Classification Report:\n", classification_report(ytest, pred))
```

```python
from sklearn.datasets import load_diabetes
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
data = load_diabetes()
X = data.data
y = data.target
Xtr, Xtest, ytr, ytest = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
Xtr_scaled = scaler.fit_transform(Xtr)
Xtest_scaled = scaler.transform(Xtest)
KnnReg=KNeighborsRegressor(n_neighbors=4)
KnnReg.fit(Xtr_scaled, ytr)
pred=KnnReg.predict(Xtest_scaled)
print("Mean Squared Error:", mean_squared_error(ytest, pred))
print("R^2 Score:", r2_score(ytest, pred))
```


## Key Parameters

- **n_neighbors**

  - number of n nearest neighboring points to consider

- **weights**

  - `'uniform'` → all neighbors contribute equally  
  - `'distance'` → closer points have higher influence  

  Theooretically we assume uniform voting, but weighted voting can improve performance.

- **metric**

  Distance function used:

  - `'euclidean'` (default)  
  - `'manhattan'`  
  - `'minkowski'`  



- **p** (for Minkowski)

  - p = 1 → Manhattan  
  - p = 2 → Euclidean  

- **algorithm**
     Specifies the method used to search for nearest neighbors.

  Options:
    `'auto'`, `'ball_tree'`, `'kd_tree'`, `'brute'`
