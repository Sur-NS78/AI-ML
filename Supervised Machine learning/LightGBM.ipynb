# Light Gradient Boosting Machine



Another gradient boosting framework developed to efficiently handle large-scale data and high-dimensional features. 
(Refer to the gradient boosting explaination here : [XGBoost](XGBoost.ipynb))

As the Base method is Gradient Boosting the equation for final prediction looks like:


$$
\hat{y} = \sum_{k=1}^{K} f_k(x)
$$

where each tree contributes to correcting the residuals of the previous trees.


Along with this Gradient Boosting structure the model also has some algorithmic and system-level improvements for fastrer training, more memory-efficiency and scalability.


### Histogram-based learning

Instead of checking every possible split value like RF, LightGBM groups continuous feature values into a fixed number of bins(specific ranges) and evaluates the splits only at boundaries of these bins. 

This helps lower computation cost while keeping performance close to exact methods.



### Leaf-wise tree growth (with depth constraint)

LightGBM uses a leaf-wise (best-first) growth strategy by identifying the leaf that gives the maximum reduction in loss and split only that leaf, after which it continues this process iteratively  

This helps reduces loss more aggressively and often leads for imporved accuracy.

However, there is a overfitting risk which is addressed by setting parameters such as `max_depth`  and  `num_leaves` 
to limit tree complexity.



### Gradient-based One-Side Sampling (GOSS)

The algorithm speeds up training by selecting only a subset of data points when computing splits. It keeps data points with large gradients(big error) and a random subset of points with small gradients ensuring that the model focuses on difficult samples while reducing computation.


### Exclusive Feature Bundling (EFB)

In many datasets(primarily sparse ones) features are often mutually exclusive (rarely take non-zero values at the same time).

This model combines such features into a single feature without losing information giving a reduction of the number of features , memory usage and computation time.

## Objective function

In the LGBM we minimize a regularized objective:

$$
\text{Objective} = \sum_{i=1}^{n} l(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
$$

- $l(y_i, \hat{y}_i)$ → loss function (can be MSE, log loss etc.)  
- $\Omega(f_k)$ → regularization term controlling tree complexity  

which is similar to other gradient boosting frameworks, ensuring a balance between accuracy and generalisation.


## The training process



LightGBM trains the model in an iterative manner using the following steps:

1. initial predictions
2. compute gradients (residuals) based on current predictions  
3. apply GOSS to select important samples  
4. apply EFB to reduce feature dimensionality  
5. build histograms for feature values  
6. choose the best split using histogram bins  
7. grow the tree using a leaf-wise strategy  
8. update predictions by adding the new tree  
9. repeat until convergence (reach minimum loss) or stopping criteria
